```

notas = []
while True:
    n = float(input("Digite uma nota (ou um n√∫mero negativo para sair): "))
    if n < 0:
        break
    notas.append(n)

if notas:
    media = sum(notas) / len(notas)
    print(f"M√©dia das notas: {media:.2f}")
else:
    print("Nenhuma nota foi digitada.")
```

```

pares, impares = [], []

while True:
    n = int(input("Digite um n√∫mero positivo (ou negativo para sair): "))
    if n < 0:
        break
    if n % 2 == 0:
        pares.append(n)
    else:
        impares.append(n)

total = len(pares) + len(impares)
if total > 0:
    p_pares = (len(pares) / total) * 100
    p_impares = (len(impares) / total) * 100
    print(f"Porcentagem de pares: {p_pares:.1f}%")
    print(f"Porcentagem de √≠mpares: {p_impares:.1f}%")
else:
    print("Nenhum n√∫mero v√°lido foi digitado.")
```

```

pessoas = []

for i in range(5):
    altura = float(input(f"Altura da {i+1}¬™ pessoa (m): "))
    peso = float(input(f"Peso da {i+1}¬™ pessoa (kg): "))
    pessoas.append((altura, peso))

iguais = any(pessoas.count(p) > 1 for p in pessoas)
if iguais:
    print("H√° pelo menos duas pessoas com mesma altura e peso.")
else:
    print("N√£o h√° pessoas com mesma altura e peso.")
```

```

pessoas = []

while True:
    nome = input("Digite o nome (ou pressione Enter para sair): ").strip()
    if not nome:
        break
    idade = int(input(f"Digite a idade de {nome}: "))
    pessoas.append((nome, idade))

if pessoas:
    mais_velho = max(pessoas, key=lambda p: p[1])
    print(f"A pessoa mais velha √© {mais_velho[0]} com {mais_velho[1]} anos.")
else:
    print("Nenhuma pessoa foi cadastrada.")
```

```

animes = {
    "Naruto": 220,
    "Jujutsu Kaisen": 47,
    "Dragon Ball Z": 291,
    "Death Note": 37,
    "Fullmetal Alchemist": 64,
    "Evangelion": 26,
    "Berserk": 25,
    "Code Geass": 50,
    "Akame ga Kill!": 24,
    "Elfen Lied": 13
}

total_ep = sum(animes.values())
print("Total de epis√≥dios:", total_ep)

percentuais = {anime: (ep / total_ep) * 100 for anime, ep in animes.items()}

print("\nPercentuais (ordem decrescente):")
for nome, perc in sorted(percentuais.items(), key=lambda x: x[1], reverse=True):
    print(f"{nome}: {perc:.2f}%")
```

```

bandas = {
    "Gangrena Gasosa": [
        {"√°lbum": "Gente Ruim S√≥ Manda Lembran√ßa Pra Quem N√£o Presta", "ano": 2019},
        {"√°lbum": "Se Deus √â 10,Satan√°s √â 666", "ano": 1999},
        {"√°lbum": "Welcome to the Macumba", "ano": 1993}
    ],
    "Rog√©rio Skylab": [
        {"√°lbum": "Nas Portas do Cu", "ano": 2023},
        {"√°lbum": "Cr√≠tica da Faculdade do Cu", "ano": 2021},
        {"√°lbum": "Cosmos", "ano": 2019}
    ],
    "Garotos Podres": [
        {"√°lbum": "Can√ß√µes para Ninar", "ano": 2003},
        {"√°lbum": "Com a Corda Toda", "ano": 1997},
        {"√°lbum": "Rock de Sub√∫rbio", "ano": 1995}
    ],
    "Massacration": [
        {"√°lbum": "Live Metal Espancation", "ano": 2017},
        {"√°lbum": "Good Blood Headbanguers", "ano": 2009},
        {"√°lbum": "Gates of Metal Fried Chicken of Death", "ano": 2005}
    ],
    "Raimundos": [
        {"√°lbum": "Cantigas de Roda", "ano": 2014},
        {"√°lbum": "Roda Viva", "ano": 2006},
        {"√°lbum": "√âramos 4", "ano": 2001}
    ]
}

# 1. √Ålbum mais antigo
mais_antigo = min((a for albuns in bandas.values() for a in albuns), key=lambda x: x["ano"])
print("√Ålbum mais antigo:", mais_antigo["√°lbum"], "-", mais_antigo["ano"])

# 2. Banda com maior intervalo entre √°lbuns
maior_intervalo = ("", 0)
for banda, albuns in bandas.items():
    anos = [a["ano"] for a in albuns]
    intervalo = max(anos) - min(anos)
    if intervalo > maior_intervalo[1]:
        maior_intervalo = (banda, intervalo)
print("Banda com maior intervalo:", maior_intervalo[0], "-", maior_intervalo[1], "anos")

# 3. Todos os √°lbuns do mais novo para o mais antigo
todos = [(a["ano"], b, a["√°lbum"]) for b, albuns in bandas.items() for a in albuns]
for ano, banda, album in sorted(todos, reverse=True):
    print(f"{ano} - {banda}: {album}")
```
